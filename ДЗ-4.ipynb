О непрерывной равномерно распределенной случайной величине `B` известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины `B` и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

Дисперсия равномерно распределенной случайно величины рассчитывается по формуле:  
$D(X) = \dfrac{(b - a)^2}{12}$  
подставим значение диспресии и выразим b:  
$0.2 = \dfrac{(b - a)^2}{12}$  
$2.4 = {(b - a)^2}$  
$ \sqrt{2.4} = {b - a}$  
$b = 1.549 + a$  

```python
import math
a = 0.5
b = math.sqrt(2.4) + a
m = (a+b)/2
print(f'Ответ: Правая граница величины B равна: {b}\n среднее значение равно: {m}')
```

На сколько сигм (средних квадратических отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой `M(X) = 178` и `D(X) = 25`?

сигма(среднее квадратическое отклонение) в нашем случае равно корню квадратному из значения дисперсии

$$\sigma = \sqrt{D(X)}$$
$$a = M(X)$$

далее, зная, что  
$a + n*\sigma = 190$  
$178 + n*5 = 190$
выразим n и найдем ско  

```python
n = (190 - 178)/5
n
```

__Задача 3__

Рост взрослого населения города `X` имеет нормальное распределение, причем, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
1. больше 182 см?
2. больше 190 см?
3. от 166 см до 190 см?
4. от 166 см до 182 см?
5. от 158 см до 190 см?
6. не выше 150 см или не ниже 190 см?
7. не выше 150 см или не ниже 198 см?
8. ниже 166 см?

Т.к. у нас распрделение нормальное, то плотность распределения задается формулой:  
$$f(x) = \dfrac{1}{\sigma \sqrt{2 \pi}} \: e^{- \dfrac{(x - a)^2}{2 \sigma^2}}.$$

Чтобы найти вероятность для каждого пункта найдем интеграл от плотности распределения вероятности в необходимом диапозоне  

$$1. P = \int_{182}^{inf} f(x)dx$$  

$$2. P = \int_{190}^{inf} f(x)dx$$

$$3. P = \int_{166}^{190} f(x)dx$$

$$4. P = \int_{166}^{182} f(x)dx$$

$$5. P = \int_{158}^{190} f(x)dx$$

$$6. P = \int_{-inf}^{150} f(x)dx + \int_{190}^{inf} f(x)dx$$

$$7. P = \int_{-inf}^{150} f(x)dx + \int_{198}^{inf} f(x)dx$$

$$8. P = \int_{-inf}^{166} f(x)dx$$


```python
import numpy as np
from scipy.integrate import quad
def pdf_normal(x):
    return np.exp(- ((x - 174) ** 2) / (2 * 8 ** 2)) / (8 * np.sqrt(2 * np.pi))

a_1 = 182
b_1 = np.inf
p_1 = quad(pdf_normal, a_1, b_1)[0]

a_2 = 190
b_2 = np.inf
p_2 = quad(pdf_normal, a_2, b_2)[0]

a_3 = 166
b_3 = 190
p_3 = quad(pdf_normal, a_3, b_3)[0]

a_4 = 166
b_4 = 182
p_4 = quad(pdf_normal, a_4, b_4)[0]

a_5 = 158
b_5 = 190
p_5 = quad(pdf_normal, a_5, b_5)[0]

a_6_1 = np.NINF
b_6_1 = 150
a_6_2 = 190
b_6_2 = np.inf
p_6_1 = quad(pdf_normal, a_6_1, b_6_1)[0]
p_6_2 = quad(pdf_normal, a_6_2, b_6_2)[0]
p_6 = p_6_1 + p_6_2

a_7_1 = np.NINF
b_7_1 = 150
a_7_2 = 198
b_7_2 = np.inf
p_7_1 = quad(pdf_normal, a_7_1, b_7_1)[0]
p_7_2 = quad(pdf_normal, a_7_2, b_7_2)[0]
p_7 = p_7_1 + p_7_2

a_8 = np.NINF
b_8 = 166
p_8 = quad(pdf_normal, a_8, b_8)[0]
p_8 = p_8
```

```python
print(f'Ответ:')
print(f'1.Вероятность того, что человек имеет рост больше 182 см = {p_1}') 
print(f'2.Вероятность того, что человек имеет рост больше 190 см = {p_2}') 
print(f'3.Вероятность того, что человек имеет рост от 166 см до 190 см = {p_3}') 
print(f'4.Вероятность того, что человек имеет рост от 166 см до 182 см = {p_4}') 
print(f'5.Вероятность того, что человек имеет рост от 158 см до 190 см = {p_5}') 
print(f'6.Вероятность того, что человек имеет рост не выше 150 см или не ниже 190 см = {p_6}') 
print(f'7.Вероятность того, что человек имеет рост не выше 150 см или не ниже 198 см = {p_7}') 
print(f'8.Вероятность того, что человек имеет рост ниже 166 см = {p_8}') 

#p_1, p_2, p_3, p_4, p_5, p_6, p_7, p_8
```
