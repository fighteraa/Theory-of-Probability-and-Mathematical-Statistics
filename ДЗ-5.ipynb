__Задача 1__

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя уровень значимости 0.05 и правостороннюю критическую область, проверить данную гипотезу при условии, что в выборке из 100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.

Гипотеза $H_0$  - средний диаметр шарика 17 мм.  
Гипотеза $H_1$  - средний диаметр шарика не 17 мм.  

Зададим функцию от выборки $T$. В качестве такой функции возьмём выборочное среднее:

$$T(X) = \dfrac{1}{n} \displaystyle\sum_{i = 1}^n x_i,$$

Уровень значимости $\alpha = 0.025$.

Поскольку наша гипотеза заключается в том, что рассматриваемая случайная величина имеет нормальное распределение с параметрами $a = 17$ и $\sigma = 2$, случайная величина $T$ будет также иметь нормальное распределение с параметрами $a = 9.8$ и $\sigma = \dfrac{2}{\sqrt{n}}$.

Критическую область представляет собой множество значений $\Omega_\alpha$ случайной величины $T$, такое, что вероятность для случайной величины $T$ попасть в множество $\Omega_\alpha$, равно $\alpha$. Поскольку величина $T$ распределена нормально, можно воспользоваться правилом двух сигм: значения величины $T$ с вероятностью 0.95 попадают в интервал

$$\left( 17 - \dfrac{4}{\sqrt{n}}, 17 + \dfrac{4}{\sqrt{n}} \right).$$

Поэтому для нашего выбора $\alpha$ = 0.025 подойдёт двусторонняя критическая область

$$\left( 17 + \dfrac{4}{\sqrt{n}}, \infty \right).$$
Размер выборки у нас равен 100 и правосторонний интервал, то критическая область будет иметь вид

$$\left( 17.4, \infty \right).$$

Если это значение попало в критическую область, то гипотеза $H_0$ отвергается в пользу гипотезы $H_1$. Если же это значение не попадает в критическую область, то гипотеза $H_0$ остаётся в силе.

Ответ: Т.к. в результате эксперимента среднее значение составило 17.5, это начит, что мы попали в критическую облать. Это значит, что гипотеза $H_0$  - средний диаметр шарика 17 мм, неверна. Верна гипотеза $H_1$  - средний диаметр шарика не 17 мм

__Задача 2__

Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочное среднее равно 80, а объём равен 256.

У нас случай, когда мы не знаем дисперию нормально распределённой случайной величины. 

Воспользуемсяс __t-распределением__.

$$
P \left( 
\overline{X} - t_{1 - \alpha/2, \: n - 1} \cdot \dfrac{S}{\sqrt{n}} 
\leq M(X) \leq
\overline{X} + t_{1 - \alpha/2, \: n - 1} \cdot \dfrac{S}{\sqrt{n}} 
\right) = 1 - \alpha.
$$

где $\overline{X}$ выборочное среднее, а через $S$ — несмещённое среднее квадратическое отклонение, n=256  
Значения коэффициентов Стьюднта вычислим с помощью библиотеки scypi и подставим в формулу.

````python
import numpy as np
from scipy import stats
p = 0.95
n = 256
t = scipy.stats.t.ppf((1 + p)/2, n-1)
print(f'Коэффициент Стьюдента = {t}')

m = 16
s = 80

print(f'Ответ: Доверительный интервал {(m - t * s / np.sqrt(n), m + t * s / np.sqrt(n))}') 
````

__Задача 3__

Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
```
202, 203, 199, 197, 195, 201, 200, 204, 194, 190
```
Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?

Гипотеза $H_0$  - средний вес пачки составляет 200 г.  
Гипотеза $H_1$  - средний вес пачки составляет не 200 г.

Доверительная вероятность равна 0.99

У нас случай, когда мы не знаем дисперию нормально распределённой случайной величины. 

Воспользуемсяс __t-распределением__.

$$
P \left( 
\overline{X} - t_{1 - \alpha/2, \: n - 1} \cdot \dfrac{S}{\sqrt{n}} 
\leq M(X) \leq
\overline{X} + t_{1 - \alpha/2, \: n - 1} \cdot \dfrac{S}{\sqrt{n}} 
\right) = 1 - \alpha.
$$

где $\overline{X}$ выборочное среднее, а через $S$ — несмещённое среднее квадратическое отклонение, n=256  
Значения коэффициентов Стьюднта вычислим с помощью библиотеки scypi и подставим в формулу.

Найдем среднее значение в выборке и среднее квадратичное отклочение по формулам  
$$\overline{x} = \dfrac{1}{m} \displaystyle\sum_{i = 1}^m x_i.$$

__не смещенную дисперсию случайной величины__ считаем по формуле:

$$\sigma^2 = \dfrac{1}{m - 1} \displaystyle\sum_{i = 1}^{m} (x_i - \overline{x})^2.$$

__среднее квадратическое отклонение__ считаем по формуле:

$$\sigma = \sqrt{\dfrac{1}{m - 1} \displaystyle\sum_{i = 1}^m (x_i - \overline{x})^2}.$$

````python
import math
cookies = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])

p = 0.99
n = 10
t = scipy.stats.t.ppf((1 + p)/2, n-1)
print(f'Коэффициент Стьюдента = {t}')

m = np.mean(cookies)
summa = 0
for i in cookies:
    summa += (i - m)**2
cookies_var = summa/(len(cookies)-1)
s = math.sqrt(cookies_var)

print(f'Доверительный интервал {(m - t * s / np.sqrt(n), m + t * s / np.sqrt(n))}')
print(f'Т.к заначение 200 г попадает в доверительный интервал, принимаем решение, что гипотеза H0 верна ')
````
