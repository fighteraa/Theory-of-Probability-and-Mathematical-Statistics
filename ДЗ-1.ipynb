__Задача 2__

Группа студентов изучает $10$ различных дисциплин. Сколькими способами можно составить расписание на понедельник, если в этот день должно быть $4$ разных занятия?

```python
import itertools
data = list(itertools.permutations(range(10), 4))
print(f'Составить расписание на понедельник можно {len(data)} способами')
```

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что два определённых человека будут сидеть рядом?

$$P(A) = \frac{m}{n}$$

Мы можем из 7 человек образовать 2*6 и оставшиеся рассядутся 5!  
Число возможных вариантов сесть 7 людям 7!  


$$P = \frac{m}{n} = \frac{2*6*5!}{7!} = \frac{2}{7} $$

Бросается игральная кость. Пусть событие $A$ - появление чётного числа, событие $B$ - появление числа больше трёх. Являются ли эти события независимыми?

Из определения независимых событий следует, что два случайных события называются независимыми, если наступление одного из них не изменяет вероятность наступления другого.

Возможные исходы события А: 2, 4, 6  
Возможные исходы события В: 4, 5, 6

В связи с тем, что у нас есть одинаковые исходы 4 и 6 у каждого из событий, они влияют друг на друга и не могут быть независимыми событиями.

__Задача 6__

Допустим, имеется некоторая очень редкая болезнь (поражает $0.1 \%$ населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет $99 \%$ больных этой болезнью и всего лишь в $1 \%$ случаев даёт ложный положительный ответ.

_Вопрос_: вакова вероятность, что вы действительно больны ей?

_Подсказка_: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

$$P(болен|тест+) = \frac{P(тест+|болен)*P(болен)}{P(тест+)} $$  
$$P(тест+) = {P(тест+|болен)*P(болен) + P(тест-|здоров)*P(здоров)}$$  
$$P(болен|тест+) = \frac{P(тест+|болен)*P(болен)}{P(тест+|болен)*P(болен) + P(тест-|здоров)*P(здоров)} $$  

```python
P = (0.99*0.001)/(0.99*0.001 + 0.01*0.999)
print(f'Вероятность того, что действительно болен краснухой {P}')
```
